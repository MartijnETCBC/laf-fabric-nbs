# Rendering Hebrew Text plus Linguistic Annotations

We will present a *rich hebrew text* representation of selected verses.
Here is an example: <a target="blank" href="images/Genesis_1_24.html">Genesis 1:24</a>

```python
import sys, os
import collections

base_dir = '{}/Dropbox/laf-fabric-output/etcbc4/shebanq'.format(os.path.expanduser('~'))
def infile(f): return open('{}/{}'.format(base_dir, f))
def outfile(f): return open('{}/{}'.format(base_dir, f), mode='w')
```

# Read the rich text data

There is a tab-separated file containing essentially the same info as the sql data generated by the notebook *laf2shebanq*.
We read this file and translate it to row-like data.

We also read a file with the plain text of the verses.

```python
sys.stderr.write('Reading rich text data ...')
richf = infile('word_data.tsv')
plainf = infile('verse_plain.txt')

passages = collections.defaultdict(lambda: [])
plain_text = {}

field_names = next(richf).strip().split('\t')
for (e, fn) in enumerate(field_names):
    print("{:<20} = {:>2}".format(fn, e))
for line in richf:
    line = line.strip()
    if line.startswith('!'):
        passage = line[1:]
    else:
        passages[passage].append(tuple(line.split('\t')))
richf.close()

for line in plainf:
    line = line[0:-1]
    (passage, text) = line.split('\t', 1)
    plain_text[passage] = text.replace('\\n', '\n')
plainf.close()
sys.stderr.write('Done')
```

```python
base_doc = 'http://shebanq-doc.readthedocs.org/en/latest/features/comments'
js_def_file = 'rhtext_def.js'
js_call_file = 'rhtext_call.js'
css_file = 'rhtext.css'

html_start = '''<html>
<head>
<meta http-equiv="Content-Type" content="text/html; charset=utf-8">
<style type="text/css">
{css}
</style>
<script src="file:jquery.js"></script>
<script src="file:{cjs}"></script>
</script>
</head>
<body>
'''.format(css=css, cjs=js_def_file)

def js_toggle(cls):
    return '''
function set_toggle_{cls}(init) {{
    $("#toggle_{cls}").attr('checked', init);
    if (!init) {{
        $('.{cls}').each(function () {{
            $( this ).toggle();
        }});
    }}
    $("#toggle_{cls}").change(function() {{
        $('.{cls}').each(function () {{
            $( this ).toggle();
        }});
        ajax(save_url + '?tg=toggle_{cls}', ['toggle_{cls}'], 'curviewlnk')
    }});
}}
'''.format(cls=cls)

html_end = '''
</body>
</html>
'''

html_selector = '''<p class="sel">
    <input type="checkbox" id="toggle_txt_p"/>plain text - 
    <input type="checkbox" id="toggle_txt_il"/>rich text
</p>
'''
html_vhead = '''<h2 class="vlabel">{}</h2>\n'''

html_vstart = '''<div class="txt_il">'''
html_pstart = '''<div class="txt_p">'''

html_vend = '''</div>'''
html_pend = '''</div>'''


htext_tpl = '''<table class="il c">
'''
for f in html_field_specs:
    cls = f['class']
    border_template = '' if f['border'] == '' else ''' {{{}}}'''.format(f['border'])
    htext_tpl += '''\t<tr class="il {cls}"><td class="il {cls}{bt}">'''.format(cls=cls, bt=border_template)
    for comp in f['comps']:
        if type(comp) is tuple:
            if len(comp) == 4:
                htext_tpl += '''<span class="{}">{{{}}}</span>'''.format(comp[0], comp[1])
            else:
                htext_tpl += '''<span class="{} {}_{}">{{{}}}</span>'''.format(comp[0], cls, comp[1], comp[2])
        else:
            htext_tpl += comp
    htext_tpl += '''</td></tr>\n'''
htext_tpl += '''</table>'''

hlegend_tpl = '''<table class="il">
'''
js = ''
for cls in ('txt_il', 'txt_p'):
    js += js_toggle(cls)

for f in html_field_specs:
    cls = f['class']
    hlegend_tpl += '''\t<tr class="il l_{cls}"><td class="c l_{cls}"><input
    type="checkbox" id="toggle_{cls}"/></td><td
    class="il l_{cls}">'''.format(cls=cls)
    js += js_toggle(cls)
    for comp in f['comps']:
        if type(comp) is tuple:
            this_cls = None
            if len(comp) == 4:
                this_cls = comp[0]
                url = '{{base_doc}}/{}.html'.format(comp[3])
                hlegend_tpl += '''<a target="_blank" href="{}"><span class="l_{}">{}</span></a>'''.format(
                    url, this_cls, comp[2],
                )
            else:
                this_cls = '{}_{}'.format(cls, comp[1])
                url = '{{base_doc}}/{}.html'.format(comp[4])
                hlegend_tpl += '''<input type="checkbox" id="toggle_{cls}"/><a target="_blank" href="{url}"><span
    class="{c} l_{cls}">{lab}</span></a>'''.format(
                    c=comp[0], cls=this_cls, lab=comp[3], url=url
                )
            if cls != this_cls:
                js += js_toggle(this_cls)
        else:
            hlegend_tpl += comp
    hlegend_tpl += '''</td></tr>\n'''
hlegend_tpl += '''</table>
'''

def h_esc(material, fill=True):
    material = material.replace(
        '&', '&amp;').replace('<', '&lt;').replace('>', '&gt;').replace('"', '&quot').replace(
        "'", '&apos;').replace('\\n', '\n')
    if fill:
        if material == '': material = '&nbsp;'
    return material
```

```python
def generate_fixed_stuff():
    for (fname, text) in (
        (css_file, css),
        (js_def_file, js),
        ('rhtext_fields.tsv', '{}\n'.format('\t'.join(field_names))),
        ('rhtext.tpl', htext_tpl),
        ('rhtext_legend.tpl', hlegend_tpl),
    ):
        with outfile(fname) as fh: fh.write(text)
    
def show_verse(passage):
    verse_data = passages[passage]
    verse_text = plain_text[passage]
    word_html = []
    for word_data in verse_data:
        word_info = dict(
            (
                field_names[i], h_esc(word_data[i], fill=not field_names[i].endswith('_border'))
            ) for i in range(len(word_data))
        )
        word_html.append(htext_tpl.format(**word_info))
    return '{}{}{}{}{}{}{}'.format(
        html_vhead.format(passage),
        html_pstart,
        verse_text,
        html_pend,
        html_vstart,
        ''.join(wh for wh in word_html),
        html_vend,
    )

def show_passages(ps):
    html = '{}{}{}{}{}'.format(
        html_start,
        html_selector,
        html_vstart + hlegend_tpl.format(base_doc=base_doc) + html_vend,
        ''.join(show_verse(passage) for passage in ps),
        html_end,
    )
    h = outfile('{}.html'.format(ps[0].replace(':','_')))
    h.write(html)
    h.close()
```

```python
generate_fixed_stuff()
show_passages(['{} {}:{}'.format('Genesis', 1, v) for v in range(24, 31)])
```

```python
q_colors = '''
ff0000
ff6688
ffcc66
ffff00
00ff00
ccff66
66ffcc
00ffff
0000ff
66ccff
cc66ff
ff00ff
'''.strip().split()

(nrows, ncols) = (3, 4)

def ccell(q,c): return '\t\t<td class="cc {q}" style="background-color: #{c}">&nbsp;</td>'.format(q=q,c=q_colors[c])
def crow(q,r): return '\t<tr>\n{}\n\t</tr>'.format('\n'.join(ccell(q,c) for c in range(r * ncols, (r + 1) * ncols)))
def ctable(q): return '<table class="picker" id="picker_{q}">\n{cs}\n</table>\n'.format(q=q, cs='\n'.join(crow(q,r) for r in range(nrows)))
def qsel(q): return '<table class="picked"><tr><td class="cc_sel" id="sel_{q}">&nbsp;</td></tr></table>\n'.format(q=q)
def js(q, initc): return '''
$('#picker_{q}').hide()
$('#sel_{q}').click(function() {{
    $('#picker_{q}').show()
}})
$('.cc.{q}').click(function() {{
    $('#picker_{q}').hide()
    $('#sel_{q}').css('background-color', $(this).css('background-color'))
}})
{cm}$('#sel_{q}').css('background-color', '#{ic}')
'''.format(q=q, ic=initc, cm='' if initc in q_colors else '//')

def picker(q,initc): return '{s}{p}<script type="text/javascript">{j}</script>\n'.format(s=qsel(q), p=ctable(q), j=js(q, initc))


htmlqc = '''
<html>
<head>
<style type="text/css">
table.picked {{
    display: inline-table;
    padding: 0;
    margin: 0;
    border: 0;
}}
table.picker {{
    display: inline-table;
    padding: 0;
    margin: 0;
    border: 0;
}}
td.cc {{
    width: 1em;
    height: 0.5em;
    line-height: 0.5em;
    padding: 0;
    margin: 0;
    border: 0;
    text-align: center;
}}
td.cc_sel {{
    width: 1em;
    height: 0.5em;
    line-height: 0.5em;
    padding: 0;
    margin: 0;
    border: 1pt solid #aaaaaa;
    text-align: center;
}}
</style>
<script src="file:jquery.js"></script>
</head>
<body>
<div>
{p1}
</div>
<div>
{p2}
</div>
<div>
{p3}
</div>
</body>
</html>
'''

with outfile('qcolors.html') as of: of.write(htmlqc.format(
    p1=picker("a", 'ff0000'),
    p2=picker("b", '0000ff'), 
    p3=picker("c", ''), 
))
```
