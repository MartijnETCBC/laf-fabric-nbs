# Common functions

```python
def get_set(monads):
    monad_set = set()
    for rn in monads.split(','):
        bnds = rn.split('-', 1)
        if len(bnds) == 1:
            monad_set.add(int(bnds[0]))
        else: 
            monad_set |= set(range(int(bnds[0]), int(bnds[1]) + 1))
    return frozenset(monad_set)

def ranges(monadset):
    result = []
    cur_start = None
    cur_end = None
    for i in sorted(monadset):
        if cur_start == None:
            cur_start = i
            cur_end = i
        else:
            if i == cur_end + 1:
                cur_end += 1
            else:
                result.append((cur_start, cur_end))
                cur_start = i
                cur_end = i
    if cur_start != None:
        result.append((cur_start, cur_end))
    return result
```

# Passage index

When we have found our objects, we want to indicate where they occur in the bible. In order to specify the passage of a node, we have to now in what verse a node occurs. In the next code cell we create a mapping from nodes of type sentence, clause, etc to nodes of type verse. From a verse node we can read off the passage information.

Conversely, we also construct an index from verses to nodes: given a verse, we make a list of all nodes belonging to that verse, in the canonical order.

```python
msg("Making passage index ...")
target_types = {
    'sentence', # 'sentence_atom', 
    'clause', # 'clause_atom', 
    'phrase', # 'phrase_atom', 
    'subphrase',
    'word',
}

passage_verse = {}
object_index = collections.defaultdict(lambda: set())
verse_monads = {}
verse_node = {}

for n in NN():
    otype = F.otype.v(n)
    if otype == 'verse':
        verse_monads[n] = get_set(F.monads.v(n))
        passage_verse[(F.book.v(n), int(F.chapter.v(n)), int(F.verse.v(n)))] = n
    elif otype in target_types:
        for m in get_set(F.monads.v(n)):
            object_index[m].add(n)
for v in verse_monads:
    vobjects = set()
    for m in verse_monads[v]: vobjects |= object_index[m]
    verse_node[v] = tuple(NN(nodes=vobjects))
msg("Passage index created for {} verses".format(len(verse_node)))
```

# Check

We need to verify that each word occurrence does not occur in more than one object of the same type.

```python
msg("Checking whether words belong to multiple objects of the same type ...")
word_index = collections.defaultdict(lambda: collections.defaultdict(lambda: set()))
for n in NN():
    otype = F.otype.v(n)
    if otype in target_types:
        for m in get_set(F.monads.v(n)):
            word_index[otype][m].add(n)
errors = 0
for otype in sorted(word_index):
    good = 0
    for monad in sorted(word_index[otype]):
        lt = len(word_index[otype][m])
        if lt <= 1: good += 1
        else:
            print("Monad {} belongs to {} {}s".format(m, lt, otype))
            errors += 1
    print("{} monads occur in only one {}".format(good, otype))
msg("End checking, {} errors".format(errors))
```

```python
html_start = '''<html>
<head>
<meta http-equiv="Content-Type" content="text/html; charset=utf-8">
<style type="text/css">
h2.vlabel {
    text-align: right;
    font-family: Ezra SIL, SBL Hebrew, Verdana, sans-serif;
    font-size: 18pt;
    line-height: 24pt;
    font-weight: bold;
}
table.inter {
    display: inline-table;
    empty-cells: show;
    margin-top: 1pt;
    margin-bottom: 1pt;
    margin-left: 0pt;
    margin-right: 0pt;
    padding: 0pt;
    border-collapse: separate;
    border-top: 2pt solid gray;
    border-bottom: 2pt solid gray;
    border-left: 1pt solid #dddddd;;
    border-right: 1pt solid #dddddd;
}
tr.layer {
    padding: 0pt;
    margin: 0pt;
}
.heb {
    text-align: right;
    font-family: Ezra SIL, SBL Hebrew, Verdana, sans-serif;
    font-size: 18pt;
    line-height: 24pt;
    height: 30pt;
    direction: rtl;
}
.tran {
    text-align: left;
    font-family: Menlo;
    font-size: 9pt;
    line-height: 11pt;
    direction: ltr;
}
.vlex {
    text-align: right;
    font-family: Ezra SIL, SBL Hebrew, Verdana, sans-serif;
    font-size: 18pt;
    line-height: 24pt;
    color: #0000ff;
    direction: ltr;
}
.lex {
    text-align: left;
    font-family: Menlo;
    font-size: 9pt;
    line-height: 11pt;
    color: #0000ff;
    direction: ltr;
}
.gloss {
    text-align: left;
    font-family: Verdana;
    font-size: 10pt;
    line-height: 11pt;
    color: #bb0000;
    direction: ltr;
    width: 8em;
    height: 24pt;
}
.word {
    text-align: center;
    font-family: Verdana;
    font-size: 9pt;
    line-height: 10pt;
    color: #666666;
    direction: ltr;
    background-color: #eeeeff;
    padding: 0pt;
    margin: 0pt;
}
.wordi {
    text-align: right;
    font-family: Verdana;
    font-size: 12pt;
    line-height: 14pt;
    color: #000000;
    direction: ltr;
    background-color: #eeeeff;
    padding: 0pt;
    margin: 0pt;
}
.subphrase {
    text-align: center;
    font-family: Verdana;
    font-size: 9pt;
    line-height: 10pt;
    color: #666666;
    direction: ltr;
    padding: 0pt;
    margin: 0pt;
}
.subphrasei {
    text-align: right;
    font-family: Verdana;
    font-size: 12pt;
    line-height: 14pt;
    color: #000000;
    direction: ltr;
    padding: 0pt;
    margin: 0pt;
}
.subphrasey {
    background-color: #ccddff;
}
.phrase {
    text-align: center;
    font-family: Verdana;
    font-size: 9pt;
    line-height: 10pt;
    color: #666666;
    direction: ltr;
    background-color: #99ffff;
    padding: 0pt;
    margin: 0pt;
}
.phrasei {
    text-align: right;
    font-family: Verdana;
    font-size: 12pt;
    line-height: 14pt;
    color: #000000;
    direction: ltr;
    background-color: #99ffff;
    padding: 0pt;
    margin: 0pt;
}
.clause {
    text-align: center;
    font-family: Verdana;
    font-size: 9pt;
    line-height: 10pt;
    color: #666666;
    direction: ltr;
    background-color: #aaffaa;
    padding: 0pt;
    margin: 0pt;
}
.clausei {
    text-align: right;
    font-family: Verdana;
    font-size: 12pt;
    line-height: 14pt;
    color: #000000;
    direction: ltr;
    background-color: #aaffaa;
    padding: 0pt;
    margin: 0pt;
}
.sentence {
    text-align: center;
    font-family: Verdana;
    font-size: 9pt;
    line-height: 10pt;
    color: #666666;
    direction: ltr;
    background-color: #eeff33;
    padding: 0pt;
    margin: 0pt;
}
.rr {
    border-right: 4pt groove silver;
}
.r {
    border-right: 4pt ridge gold;
}
.ll {
    border-left: 4pt groove silver;
}
.l {
    border-left: 4pt ridge gold;
}
.line {
    direction: rtl;
}
</style>
</head>
<body>
'''
html_end = '''
</body>
</html>
'''

html_vhead = '''<h2 class="vlabel">{} {}: {}</h2>\n'''

html_vstart = '''<p class="line">'''
html_vend = '''</p>'''

html_fields = '''
    heb vlex tran lex gloss word wordi subphrase subphrasei phrase phrasei clause clausei sentence
'''.strip().split()

html_labels = '''
text hebrew
vocalised lexeme
text transcription
lexeme transcription
gloss
monad number
part-of-speech
subphrase number
subphrase relation
phrase number
phrase function & type
clause number
clause type & domain
sentence number
'''.strip().split('\n')

#html_word = '''<table class="inter">{}
#</table>'''.format(
#    '\n\t'.join('<tr class="layer"><td class="{x}">{{{x}}}</td></tr>'.format(x=f) for f in html_fields)
#)

def html_word(word, border_specs):
    return '''<table class="inter">{}
</table>'''.format(
    '\n\t'.join('<tr class="layer"><td class="{f}{b}">{d}</td></tr>'.format(
        f=f, b=border_specs.get(f, ''), d=word[f]
    ) for f in html_fields)
)

def html_label():
    return '''{}<table class="inter">{}
</table>{}'''.format(
    html_vstart,
    '\n\t'.join('<tr class="layer"><th class="{f}">{l}</th></tr>'.format(
        l=html_labels[i], f=html_fields[i]) for i in range(len(html_labels))),
    html_vend,
)


def h_esc(material):
    material = material.replace(
        '&', '&amp;').replace('<', '&lt;').replace('>', '&gt;').replace('"', '&quot').replace("'", '&apos;')
    if material == '': material = '&nbsp;'
    return material

def get_word_info(ob): return '{}'.format(h_esc(F.sp.v(ob)))
def get_subphrase_info(ob):
    mat = F.rela.v(ob)
    return '{}'.format(h_esc(mat if mat != None else ''))
def get_phrase_info(ob): return '{}&nbsp;{}'.format(h_esc(F.function.v(ob)), h_esc(F.typ.v(ob)))
def get_clause_info(ob): return '{}&nbsp;{}'.format(h_esc(F.typ.v(ob)), h_esc(F.txt.v(ob)))
    
info_methods = {
    'word': get_word_info,
    'subphrase': get_subphrase_info,
    'phrase': get_phrase_info,
    'clause': get_clause_info,
}

def show_verse(passage, debug=False):
    verse = passage_verse[passage]
    (verse_startm, verse_endm) = tuple(int(x) for x in F.monads.v(verse).split('-'))
    objects = verse_node[verse]
    words = []
    word_index = {}
    border_specs = {}
    wn = 0
    object_ranges = collections.defaultdict(lambda: collections.defaultdict(lambda: []))
    for n in objects:
        otype = F.otype.v(n)
        if otype == 'word':
            word_index[int(F.monads.v(n))] = wn
            words.append(dict(
                heb=h_esc(F.g_word_utf8.v(n) + F.trailer_utf8.v(n)),
                vlex=h_esc(F.g_entry_heb.v(n)),
                tran=h_esc(F.g_word.v(n)),
                lex=h_esc(F.lex.v(n)),
                gloss=h_esc(F.gloss.v(n)),
                word=h_esc(F.monads.v(n)),
                wordi=get_word_info(n),
                subphrase='&nbsp;',
                subphrasei='&nbsp;',
                phrase='&nbsp;',
                phrasei='&nbsp;',
                clause='&nbsp;',
                clausei='&nbsp;',
                sentence='&nbsp;',
            ))
            border_specs[wn] = collections.defaultdict(lambda: '')
            wn += 1
        elif otype in {'subphrase', 'phrase', 'clause', 'sentence'}:
            object_ranges[otype][n] = ranges(get_set(F.monads.v(n)))
    nwords = len(words)
    if debug:
        print("{} {}:{} ({} words) {{{}}}".format(passage[0], passage[1], passage[2], nwords, F.monads.v(verse)))
#        for ob in objects:
#                print("\t{} {} {{{}}}".format(otype, F.number.v(ob), F.monads.v(ob)))
    for otype in object_ranges:
        these_ranges = object_ranges[otype]
        get_info = info_methods.get(otype, None)
        for ob in these_ranges:
            ob_num = F.monads.v(ob) if otype == 'word' else F.number.v(ob)
            ob_num = '' if ob_num == None else ob_num
            if debug:
                print("\t{} {} {{{}}}".format(otype, ob_num, F.monads.v(ob)))
            nranges = len(these_ranges[ob]) - 1
            for (e,r) in enumerate(these_ranges[ob]):
                is_first = e == 0
                is_last = e == nranges
                right_border = 'rr' if is_first else 'r'
                left_border = 'll' if is_last else 'l'
                first_word = word_index.get(r[0], -1 if r[0] < verse_startm else nwords)
                last_word = word_index.get(r[1], -1 if r[1] < verse_startm else nwords)
                if debug:
                    print("\t\tfrom {} {{{}}} to {} {{{}}}".format(first_word, r[0], last_word, r[1]))
                layers = None
                if get_info == None:
                    layers = (otype,)
                else:
                    layers = (otype, otype + 'i')
                    words[max((first_word, 0))][otype + 'i'] = get_info(ob)
                for i in range(max((first_word, 0)), min((last_word, nwords - 1)) + 1):
                    words[i][otype] = h_esc(ob_num)
                    if otype == 'subphrase':
                        for f in layers:
                            border_specs[i][f] += ' ' + 'subphrasey'
                for f in layers:
                    if 0 <= first_word < nwords:
                        border_specs[first_word][f] += ' ' + right_border
                    else:
                        msg("{} {} has first word outside passage: {}".format(otype, ob, r[0]), withtime=False)
                    if 0 <= last_word < nwords:
                        border_specs[last_word][f] += ' ' + left_border
                    else:
                        msg("{} {} has last word outside passage: {}".format(otype, ob, r[1]), withtime=False)
    return '{}{}{}{}'.format(
        html_vhead.format(*passage),
        html_vstart,
        ''.join(html_word(words[w], border_specs[w]) for w in range(len(words))),
        html_vend,
    )

def show_passages(passages, debug=False):
    html = '{}{}{}{}'.format(
        html_label(),
        html_start,
        ''.join(show_verse(passage, debug=debug) for passage in passages),
        html_end,
    )
    h = outfile('{}_{}_{}.html'.format(*passages[0]))
    h.write(html)
    h.close()
    return HTML(html)
```
