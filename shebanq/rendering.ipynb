# Rendering Hebrew Text plus Linguistic Annotations

We will present a *rich hebrew text* representation of selected verses.

```python
import sys, os
import collections
from IPython.display import clear_output, display, HTML
```

# Read the rich text data

There is a tab-separated file containing essentially the same info as the sql data generated by the notebook *laf2shebanq*.
We read this file and translate it to row-like data.

```python
base_dir = '{}/laf-fabric-output/etcbc4/shebanq'.format(os.path.expanduser('~'))
def infile(f): return open('{}/{}'.format(base_dir, f))
def outfile(f): return open('{}/{}'.format(base_dir, f), mode='w')

sys.stderr.write('Reading rich text data ...')
richf = infile('word_data.tsv')

passages = collections.defaultdict(lambda: [])

field_names = next(richf).strip().split('\t')
for (e, fn) in enumerate(field_names):
    print("{:<20} = {:>2}".format(fn, e))
for line in richf:
    line = line.strip()
    if line.startswith('!'):
        passage = line[1:]
    else:
        passages[passage].append(tuple(line.split('\t')))
richf.close()
sys.stderr.write('Done')
```

# Composing HTML

We want to find the objects corresponding to a given passage, and we also want to display the words and their part-of-speech.

```python
html_start = '''<html>
<head>
<meta http-equiv="Content-Type" content="text/html; charset=utf-8">
<style type="text/css">
h2.vlabel {
    text-align: right;
    font-family: Ezra SIL, SBL Hebrew, Verdana, sans-serif;
    font-size: 18pt;
    line-height: 24pt;
    font-weight: bold;
}
table.inter {
    display: inline-table;
    empty-cells: show;
    margin-top: 1pt;
    margin-bottom: 1pt;
    margin-left: 0pt;
    margin-right: 0pt;
    padding: 0pt;
    border-collapse: separate;
    border-top: 2pt solid gray;
    border-bottom: 2pt solid gray;
    border-left: 1pt solid #dddddd;;
    border-right: 1pt solid #dddddd;
}
tr.layer {
    padding: 0pt;
    margin: 0pt;
}
.heb {
    text-align: right;
    font-family: Ezra SIL, SBL Hebrew, Verdana, sans-serif;
    font-size: 18pt;
    line-height: 24pt;
    height: 30pt;
    direction: rtl;
}
.trn {
    text-align: left;
    font-family: Menlo;
    font-size: 9pt;
    line-height: 11pt;
    direction: ltr;
}
.vlx {
    text-align: right;
    font-family: Ezra SIL, SBL Hebrew, Verdana, sans-serif;
    font-size: 18pt;
    line-height: 24pt;
    color: #0000ff;
    direction: ltr;
}
.lex {
    text-align: left;
    font-family: Menlo;
    font-size: 9pt;
    line-height: 11pt;
    color: #0000ff;
    direction: ltr;
}
.gls {
    text-align: center;
    font-family: Verdana;
    font-size: 10pt;
    line-height: 12pt;
    color: #bb0000;
    direction: ltr;
    width: 8em;
    height: 24pt;
}
.n {
    font-family: Verdana;
    font-size: 8pt;
    line-height: 12pt;
    color: #444444;
    direction: ltr;
}
.i {
    font-family: Verdana;
    font-size: 10pt;
    line-height: 12pt;
    color: #000000;
    direction: ltr;
}
.wd {
    text-align: right;
    background-color: #eeeeff;
    padding: 0pt;
    margin: 0pt;
}
.sp {
    text-align: right;
    padding: 0pt;
    margin: 0pt;
}
.sy {
    background-color: #ccddff;
}
.ph {
    text-align: right;
    background-color: #99ffff;
    padding: 0pt;
    margin: 0pt;
}
.cl {
    text-align: right;
    background-color: #aaffaa;
    padding: 0pt;
    margin: 0pt;
}
.sn {
    text-align: right;
    background-color: #eeff33;
    padding: 0pt;
    margin: 0pt;
}
.rr {
    border-right: 4pt groove silver;
}
.r {
    border-right: 4pt ridge gold;
}
.ll {
    border-left: 4pt groove silver;
}
.l {
    border-left: 4pt ridge gold;
}
.line {
    direction: rtl;
}
</style>
</head>
<body>
'''
html_end = '''
</body>
</html>
'''

html_vhead = '''<h2 class="vlabel">{}</h2>\n'''

html_vstart = '''<p class="line">'''
html_vend = '''</p>'''

html_field_specs = (
    {
        'border': '',
        'class': 'heb',
        'comps': (('', 'word_heb', 'text hebrew'),),
    },
    {
        'border': '',
        'class': 'vlx',
        'comps': (('', 'word_vlex', 'vocalized lexeme'),),
    },
    {
        'border': '',
        'class': 'trn',
        'comps': (('', 'word_tran', 'text transcription'),),
    },
    {
        'border': '',
        'class': 'lex',
        'comps': (('', 'word_lex', 'lexeme transcription'),),
    },
    {
        'border': '',
        'class': 'gls',
        'comps': (('', 'word_gloss', 'gloss'),),
    },
    {
        'border': '',
        'class': 'wd',
        'comps': (
            ('i', 'word_pos', 'part-of-speech'),
            '&nbsp;',
            ('n', 'word_number', 'monad number'),
        ),
    },
    {
        'border': 'subphrase_border',
        'class': 'sp',
        'comps': (
            ('i', 'subphrase_rela', 'subphrase relation'),
            '&nbsp;',
            ('n', 'subphrase_number', 'subphrase number'),
        ),
    },
    {
        'border': 'phrase_border',
        'class': 'ph',
        'comps': (
            ('i', 'phrase_function', 'phrase function'),
            '&nbsp;|&nbsp;',
            ('i', 'phrase_typ', 'phrase type'),
            '&nbsp;',
            ('n', 'phrase_number', 'phrase number'),
        ),
    },
    {
        'border': 'clause_border',
        'class': 'cl',
        'comps': (
            ('i', 'clause_typ', 'clause type'),
            '&nbsp;|&nbsp;',
            ('i', 'clause_txt', 'clause domain'),
            '&nbsp;',
            ('n', 'clause_number', 'clause number'),
        ),
    },
    {
        'border': 'sentence_border',
        'class': 'sn',
        'comps': (('n', 'sentence_number', 'sentence number'),),
    },

)

htext_tpl = '''<table class="inter">
'''
for f in html_field_specs:
    border_template = '' if f['border'] == '' else ''' {{{}}}'''.format(f['border'])
    htext_tpl += '''\t<tr class="layer"><td class="{}">'''.format(f['class'] + border_template)
    for comp in f['comps']:
        if type(comp) is tuple:
            htext_tpl += '''<span class="{}">{{{}}}</span>'''.format(comp[0], comp[1])
        else:
            htext_tpl += comp
    htext_tpl += '''</td></tr>\n'''
htext_tpl += '''</table>
'''

hlegend_tpl = '''<table class="inter">
'''
for f in html_field_specs:
    hlegend_tpl += '''\t<tr class="layer"><td class="{}">'''.format(f['class'])
    for comp in f['comps']:
        if type(comp) is tuple:
            hlegend_tpl += '''<span class="{}">{}</span>'''.format(comp[0], comp[2])
        else:
            hlegend_tpl += comp
    hlegend_tpl += '''</td></tr>\n'''
hlegend_tpl += '''</table>
'''

def h_esc(material):
    material = material.replace(
        '&', '&amp;').replace('<', '&lt;').replace('>', '&gt;').replace('"', '&quot').replace("'", '&apos;')
    if material == '': material = '&nbsp;'
    return material

def tidy(s): return s.replace('''<span class="i">&nbsp;</span>&nbsp;<span class="n">&nbsp;</span>''','&nbsp;')

def show_verse(passage):
    verse_data = passages[passage]
    word_html = []
    for word_data in verse_data:
        word_info = dict((field_names[i], h_esc(word_data[i])) for i in range(len(word_data)))
        word_html.append(tidy(htext_tpl.format(**word_info)))
    return '{}{}{}{}'.format(
        html_vhead.format(passage),
        html_vstart,
        ''.join(wh for wh in word_html),
        html_vend,
    )

def show_passages(ps):
    html = '{}{}{}{}'.format(
        html_vstart + hlegend_tpl + html_vend,
        html_start,
        ''.join(show_verse(passage) for passage in ps),
        html_end,
    )
    h = outfile('{}.html'.format(ps[0].replace(':','_')))
    h.write(html)
    h.close()
    return HTML(html)
```

```python
html = show_passages(['{} {}:{}'.format('Genesis', 1, v) for v in range(24, 31)])
display(html)
```
