# Rendering Hebrew Text plus Linguistic Annotations

We will present a *rich hebrew text* representation of selected verses.

```python
import sys, os
import collections
from IPython.display import clear_output, display, HTML
```

# Read the rich text data

There is a tab-separated file containing essentially the same info as the sql data generated by the notebook *laf2shebanq*.
We read this file and translate it to row-like data.

```python
base_dir = '{}/Dropbox/laf-fabric-output/etcbc4/shebanq'.format(os.path.expanduser('~'))
def infile(f): return open('{}/{}'.format(base_dir, f))
def outfile(f): return open('{}/{}'.format(base_dir, f), mode='w')

sys.stderr.write('Reading rich text data ...')
richf = infile('word_data.tsv')

passages = collections.defaultdict(lambda: [])

field_names = next(richf).strip().split('\t')
for (e, fn) in enumerate(field_names):
    print("{:<20} = {:>2}".format(fn, e))
for line in richf:
    line = line.strip()
    if line.startswith('!'):
        passage = line[1:]
    else:
        passages[passage].append(tuple(line.split('\t')))
richf.close()
sys.stderr.write('Done')
```

# Composing HTML

We want to find the objects corresponding to a given passage, and we also want to display the words and their part-of-speech.

```python
js_file = 'rich_hebrew_text.js'

css_constants = dict(
    font_hebrew='Ezra SIL, SBL Hebrew, Verdana, sans-serif',
    size_hebrew='18pt',
    lineh_hebrew='24pt',
    halign_hebrew='right',

    font_trans='Menlo, Courier New, Courier, monospace',
    size_trans='9pt',
    lineh_trans='11pt',
    halign_trans='left',

    font_gloss='Verdana, Arial, sans-serif',
    size_gloss='10pt',
    lineh_gloss='11pt',
    halign_gloss='center',

    font_number='Verdana, Arial, sans-serif',
    size_number='8pt',
    lineh_number='11pt',

    font_info='Verdana, Arial, sans-serif',
    size_info='10pt',
    lineh_info='11pt',

    font_vheading='Ezra SIL, SBL Hebrew, Verdana, sans-serif',
    size_vheading='18pt',
    lineh_vheading='24pt',

    cellh_hebrew='30pt',
    cellh_trans='12pt',
    cellh_gloss='24pt',
    cellh_word='12pt',
    cellh_subphrase='12pt',
    cellh_phrase='12pt',
    cellh_clause='12pt',
    cellh_sentence='12pt',

    halign_default='right',

    color_txt='#000000',
    color_lex='#0000ff',
    color_gloss='#dd0000',
    color_number='#aaaaaa',
    color_info='#000000',

    color_subphrase_no='#ffddcc',
    color_subphrase_yes='#ccddff',
    color_phrase='#99ffff',
    color_clause='#aaffaa',
    color_sentence='#eeff33',

    color_border='#aaaaaa',
    color_start='#aaaaaa',
    color_end='#aaaaaa',
    color_resume='#ff0000',
    color_suspend='#ff0000',

    kind_border_hor='2pt solid',
    kind_border_ver='1pt solid',
    kind_border_be='4pt solid',

    padding_ver='4pt',
    padding_verx='8pt',

    width_table='9em',
)
```

```python
css = '''
h2.vlabel {{
    text-align: right;
    font-family: {font_vheading};
    font-size: {size_vheading};
    line-height: {lineh_vheading};
    font-weight: bold;
}}
p.i {{
    direction: rtl;
}}
table.i {{
    display: inline-table;
    empty-cells: show;
    margin-top: 1pt;
    margin-bottom: 1pt;
    margin-left: 0pt;
    margin-right: 0pt;
    padding-left: 0pt;
    padding-right: 0pt;
    padding-top: 0pt;
    padding-bottom: 0pt;
    border-collapse: separate;
    border-spacing: 1pt 0pt;
    border-top: {kind_border_hor} {color_border};
    border-bottom: {kind_border_hor} {color_border};
}}
table.c {{
    width: {width_table};
}}
tr.i {{
    padding-left: 0pt;
    padding-right: 0pt;
    margin-left: 0pt;
    margin-right: 0pt;
}}
td.i {{
    padding-left: {padding_ver};
    padding-right: {padding_ver};
    margin-left: 0pt;
    margin-right: 0pt;
}}
td.c {{
    text-align: center;
    padding-left: {padding_ver};
    padding-right: {padding_ver};
    margin-left: 0pt;
    margin-right: 0pt;
}}
td.ht, td.l_ht {{
    text-align: {halign_hebrew};
    direction: rtl;
    height: {cellh_hebrew};
    border-left: {kind_border_ver} {color_border};
    border-right: {kind_border_ver} {color_border};
}}
td.hl, td.l_hl {{
    text-align: {halign_hebrew};
    direction: rtl;
    height: {cellh_hebrew};
    border-left: {kind_border_ver} {color_border};
    border-right: {kind_border_ver} {color_border};
}}
td.tt, td.l_tt {{
    text-align: {halign_trans};
    direction: ltr;
    height: {cellh_trans};
    border-left: {kind_border_ver} {color_border};
    border-right: {kind_border_ver} {color_border};
}}
td.tl, td.l_tl {{
    text-align: {halign_trans};
    direction: ltr;
    height: {cellh_trans};
    border-left: {kind_border_ver} {color_border};
    border-right: {kind_border_ver} {color_border};
}}
td.gl, td.l_gl {{
    text-align: {halign_gloss};
    direction: ltr;
    height: {cellh_gloss};
    border-left: {kind_border_ver} {color_border};
    border-right: {kind_border_ver} {color_border};
}}
td.wd, td.l_wd {{
    text-align: {halign_default};
    direction: ltr;
    height: {cellh_word};
    border-left: {kind_border_ver} {color_border};
    border-right: {kind_border_ver} {color_border};
    padding-left: {padding_verx};
    padding-right: {padding_verx};
}}
td.sp, td.l_sp {{
    text-align: {halign_default};
    direction: ltr;
    height: {cellh_subphrase};
    background-color: {color_subphrase_no};
    border-left: {kind_border_be} {color_subphrase_no};
    border-right: {kind_border_be} {color_subphrase_no};
}}
td.sy, td.l_sy {{
    background-color: {color_subphrase_yes};
    border-left: {kind_border_be} {color_subphrase_yes};
    border-right: {kind_border_be} {color_subphrase_yes};
}}
td.ph, td.l_ph {{
    text-align: {halign_default};
    direction: ltr;
    height: {cellh_phrase};
    background-color: {color_phrase};
    border-left: {kind_border_be} {color_phrase};
    border-right: {kind_border_be} {color_phrase};
}}
td.cl, td.l_cl {{
    text-align: {halign_default};
    direction: ltr;
    height: {cellh_clause};
    background-color: {color_clause};
    border-left: {kind_border_be} {color_clause};
    border-right: {kind_border_be} {color_clause};
}}
td.sn, td.l_sn {{
    text-align: {halign_default};
    direction: ltr;
    height: {cellh_sentence};
    background-color: {color_sentence};
    border-left: {kind_border_be} {color_sentence};
    border-right: {kind_border_be} {color_sentence};
}}
td.rr {{
    border-right: {kind_border_be} {color_start};
}}
td.r {{
    border-right: {kind_border_be} {color_resume};
}}
td.ll {{
    border-left: {kind_border_be} {color_end};
}}
td.l {{
    border-left: {kind_border_be} {color_suspend};
}}
span.ht, span.l_ht {{
    font-family: {font_hebrew};
    font-size: {size_hebrew};
    color: {color_txt};
    line-height: {lineh_hebrew};
}}
span.hl, span.l_hl {{
    font-family: {font_hebrew};
    font-size: {size_hebrew};
    color: {color_lex};
    line-height: {lineh_hebrew};
}}
span.tt, span.l_tt {{
    font-family: {font_trans};
    font-size: {size_trans};
    color: {color_txt};
    line-height: {lineh_trans};
}}
span.tl, span.l_tl {{
    font-family: {font_trans};
    font-size: {size_trans};
    color: {color_lex};
    line-height: {lineh_trans};
}}
span.gl, span.l_gl {{
    font-family: {font_gloss};
    font-size: {size_gloss};
    color: {color_gloss};
    line-height: {lineh_gloss};
}}
span.n {{
    font-family: {font_number};
    font-size: {size_number};
    color: {color_number};
    line-height: {lineh_number};
}}
span.i {{
    font-family: {font_info};
    font-size: {size_info};
    color: {color_info};
    line-height: {lineh_info};
}}
'''.format(**css_constants)
```

```python
html_start = '''<html>
<head>
<meta http-equiv="Content-Type" content="text/html; charset=utf-8">
<style type="text/css">
{css}
</style>
<script src="file:jquery.js"></script>
<script src="file:{cjs}"></script>
</script>
</head>
<body>
'''.format(css=css, cjs=js_file)

def js_toggle(cls):
    return '''
function toggle_{cls}() {{
    $("#toggle_{cls}").attr('checked', true);
    $("#toggle_{cls}").change(function() {{
        $('.{cls}').each(function () {{
            $( this ).toggle();
        }});
    }});
}}
'''.format(cls=cls)

html_end = '''
<script>
    {js_end}
</script>

</body>
</html>
'''

html_vhead = '''<h2 class="vlabel">{}</h2>\n'''

html_vstart = '''<p class="i">'''

html_vend = '''</p>'''

html_field_specs = (
    {
        'border': '',
        'class': 'ht',
        'comps': (('ht', 'word_heb', 'text כתף'),),
    },
    {
        'border': '',
        'class': 'hl',
        'comps': (('hl', 'word_vlex', 'lexeme דבר'),),
    },
    {
        'border': '',
        'class': 'tt',
        'comps': (('tt', 'word_tran', 'text'),),
    },
    {
        'border': '',
        'class': 'tl',
        'comps': (('tl', 'word_lex', 'lexeme'),),
    },
    {
        'border': '',
        'class': 'gl',
        'comps': (('gl', 'word_gloss', 'gloss'),),
    },
    {
        'border': '',
        'class': 'wd',
        'comps': (
            ('i', 'pos', 'word_pos', 'part-of-speech'),
            '&nbsp;',
            ('n', 'n', 'word_number', 'monad#'),
        ),
    },
    {
        'border': 'subphrase_border',
        'class': 'sp',
        'comps': (
            ('i', 'rela', 'subphrase_rela', 'relation'),
            '&nbsp;',
            ('n', 'n', 'subphrase_number', 'subphrase#'),
        ),
    },
    {
        'border': 'phrase_border',
        'class': 'ph',
        'comps': (
            ('i', 'fun', 'phrase_function', 'function'),
            '&nbsp;',
            ('i', 'typ', 'phrase_typ', 'type'),
            '&nbsp;',
            ('n', 'n', 'phrase_number', 'phrase#'),
        ),
    },
    {
        'border': 'clause_border',
        'class': 'cl',
        'comps': (
            ('i', 'typ', 'clause_typ', 'type'),
            '&nbsp;',
            ('i', 'dom', 'clause_txt', 'domain'),
            '&nbsp;',
            ('n', 'n', 'clause_number', 'clause#'),
        ),
    },
    {
        'border': 'sentence_border',
        'class': 'sn',
        'comps': (('n', 'n','sentence_number', 'sentence#'),),
    },

)

htext_tpl = '''<table class="i c">
'''
for f in html_field_specs:
    cls = f['class']
    border_template = '' if f['border'] == '' else ''' {{{}}}'''.format(f['border'])
    htext_tpl += '''\t<tr class="i {cls}"><td class="i {cls}{bt}">'''.format(cls=cls, bt=border_template)
    for comp in f['comps']:
        if type(comp) is tuple:
            if len(comp) == 3:
                htext_tpl += '''<span class="{}">{{{}}}</span>'''.format(comp[0], comp[1])
            else:
                htext_tpl += '''<span class="{} {}_{}">{{{}}}</span>'''.format(comp[0], cls, comp[1], comp[2])
        else:
            htext_tpl += comp
    htext_tpl += '''</td></tr>\n'''
htext_tpl += '''</table>'''

hlegend_tpl = '''<table class="i">
'''
(js, js_end) = ('', '')

for f in html_field_specs:
    cls = f['class']
    hlegend_tpl += '''\t<tr class="i l_{cls}"><td class="c l_{cls}"><input
    type="checkbox" id="toggle_{cls}"/></td><td
    class="i l_{cls}">'''.format(cls=cls)
    js += js_toggle(cls)
    js_end += '    toggle_{}()\n'.format(cls)
    for comp in f['comps']:
        if type(comp) is tuple:
            this_cls = None
            if len(comp) == 3:
                this_cls = comp[0]
                hlegend_tpl += '''<span class="l_{}">{}</span>'''.format(this_cls, comp[2])
            else:
                this_cls = '{}_{}'.format(cls, comp[1])
                hlegend_tpl += '''<input type="checkbox" id="toggle_{cls}"/><span
    class="{c} l_{cls}">{lab}</span>
'''.format(c=comp[0], cls=this_cls, lab=comp[3])
            if cls != this_cls:
                js += js_toggle(this_cls)
                js_end += '    toggle_{}()\n'.format(this_cls)
        else:
            hlegend_tpl += comp
    hlegend_tpl += '''</td></tr>\n'''
hlegend_tpl += '''</table>
'''

def h_esc(material, fill=True):
    material = material.replace(
        '&', '&amp;').replace('<', '&lt;').replace('>', '&gt;').replace('"', '&quot').replace(
        "'", '&apos;').replace('\\n', '\n')
    if fill:
        if material == '': material = '&nbsp;'
    return material

def show_verse(passage):
    verse_data = passages[passage]
    word_html = []
    for word_data in verse_data:
        word_info = dict(
            (
                field_names[i], h_esc(word_data[i], fill=not field_names[i].endswith('_border'))
            ) for i in range(len(word_data))
        )
        word_html.append(htext_tpl.format(**word_info))
    return '{}{}{}{}'.format(
        html_vhead.format(passage),
        html_vstart,
        ''.join(wh for wh in word_html),
        html_vend,
    )

def show_passages(ps):
    html = '{}{}{}{}'.format(
        html_start,
        html_vstart + hlegend_tpl + html_vend,
        ''.join(show_verse(passage) for passage in ps),
        html_end.format(js_end=js_end),
    )
    h = outfile('{}.html'.format(ps[0].replace(':','_')))
    h.write(html)
    h.close()
    j = outfile(js_file)
    j.write(js)
    j.close()
    return HTML(html)
```

```python
html = show_passages(['{} {}:{}'.format('Genesis', 1, v) for v in range(24, 31)])
display(html)
```

# Generate SQL for the rich text info

The rich text info is made available to shebanq by means of the table *words*.
Per word it contains all the information needed to generate the rich html for the individual table-elements that correspond to the words.
